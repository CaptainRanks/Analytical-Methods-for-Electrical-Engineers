### 8. Definition of Derivative

Use the definition of the derivative to find the derivatives of the following functions.

a) $f(x) = x^3$, for any $x$

b) $g(x) = 1/x$, for $x \neq 0$

c) $h(x) = \sqrt{x}$, for $x > 0$

d) $k(x) = x^{-1/2}$ for $x > 0$.

## Answer to Problem 8

Rule of Derivatives: $$ \lim_{h \rightarrow 0} \frac{f(x+h)-f(x)}{h}$$

\\
a)

$$
\lim_{h \rightarrow 0} \frac{(x+h)^3-x^3}{h} = \frac{h^3+3h^2x+3hx^2+x^3-x^3}{h}=3x^2
$$

b)

$$
\lim_{h \rightarrow 0} \frac{\frac{1}{x+h}-\frac{1}{x}}{h}= \frac{x-(x+h)}{hx(x+h)}=\frac{-1}{x^2+h}= \frac{-1}{x^2}
$$

c)

$$
\lim_{h \rightarrow 0} \frac{\sqrt{x+h}-\sqrt{x}}{h}
$$

\\
$$
= \frac{\sqrt{x+h}-\sqrt{x}}{h} *\frac{\sqrt{x+h}+\sqrt{x}}{\sqrt{x+h}+\sqrt{x}}
$$

\\
$$
= \frac{1}{\sqrt{x+h}+\sqrt{x}} = \frac{1}{2 \sqrt{x}}
$$

d)

$$
\lim_{h \rightarrow 0} \frac{\frac{1}{\sqrt{x+h}}-\frac{1}{\sqrt{x}}}{h}
$$

\\
$$
{\frac{1}{\sqrt{x+h}}*\frac{\sqrt{x}}{\sqrt{x}}-\frac{1}{\sqrt{x}}}*\frac{\sqrt{x+h}}{\sqrt{x+h}} = \frac{\sqrt{x}-\sqrt{x+h}}{\sqrt{x^2+h}}
$$

\\
$$
\frac{\sqrt{x}-\sqrt{x+h}}{\sqrt{x^2+h}}*\frac{\sqrt{x}+\sqrt{x+h}}{\sqrt{x}+\sqrt{x+h}}= \frac{-1}{\sqrt{x^2+h}*(\sqrt{x}+\sqrt{x+h})}
$$

\\
$$
\frac{-1}{\sqrt{x^2}*2\sqrt{x}} = \frac{-1}{2x\sqrt{x}}
$$

## Answers for Problem 9

Chain Rule:  $g'(f(x)) \cdot (f(x))$

\\
a) $f(x) = \frac{1}{1+x^2}= (1+x^2)^{-1}$

$$
g(x) = (f(x))^{-1} , \ f(x) = 1+x^2
$$

\\
$$
g'(x) = -(f(x))^{-2}, \ f'(x)=2x
$$

\\
$$
g'(f(x)) \cdot (f'(x))=\frac{-2x}{(x^2+1)^2}
$$

b) $h(x) = (\sin x^k)^m$ for $m,k \in \mathbb{N}$

\\
$$
g(x) = (f(x))^{m} , \ f(x) = \sin x^k , \ h(x)=x^k
$$

\\
$$
g'(x) = m(f(x))^{m-1}, \ f'(x)= cos(x^k), \ h'(x)=kx^{k-1}
$$

\\
$$
f'(h(x)) \cdot (h'(x))=kx^{k-1}cos(x^k)
$$

\\
$$
g'(f(x)) \cdot (f'(x))=m(\sin(x^k)^{m-1}kx^{k-1}cos(x^k)
$$
